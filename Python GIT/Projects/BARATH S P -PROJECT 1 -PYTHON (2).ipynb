```python
# Task 1
import random
class BrainTeaser:
    def riddle(self):
        riddles = {
            "What has to be broken before you can use it?": "egg",
            "I’m tall when I’m young, and I’m short when I’m old. What am I?": "candle",
            "What month of the year has 28 days?": "all of them",
            "What is full of holes but still holds water?": "sponge",
            "What question can you never answer yes to": "are you asleep yet?"
        }
        list_riddle = random.sample(list(riddles.items()), 3)
        count = 0
        total_questions = 0
        for question, answer in list_riddle:
            print("\nQuestion:", question)
            user_answer = input("Enter the answer: ").strip().lower()
            if user_answer == answer:
                print("Your answer is correct!")
                count += 1  
            else:
                print(f"Incorrect. The correct answer is: {answer}")
            total_questions += 1 
        if total_questions > 0:
            print(f"\nYour total score: {count}/{total_questions}")
            print(f"Your score: {(count / total_questions) * 100}%")
            print("-----------------------------------------------")
        else:
            print("No Questions were found")
        choice = input("Do you want to play again? (yes/no): ").strip().lower()
        if choice == 'yes':
            self.riddle()
        print("Thank you for playing!")
        print("--------------------------------------------------")

    def scrambled_words(self):
        scrambled_words_words = {
            'fegirar': 'giraffe',
            'ignenup': 'penguin',
            'lubremal': 'umbrella',
            'roonkaga': 'kangaroo',
            'neppailpe': 'pineapple'
        }
        list_scrambledwords = random.sample(list(scrambled_words_words.items()), 3)
        count = 0
        total_questions = 0
        for questions, answers in list_scrambledwords:
            print("Question:", questions)
            user_answer = input("Enter the answer: ").strip().lower()
            if user_answer == answers:
                print("Your answer is correct!")
                count += 1  
            else:
                print(f"Incorrect. The correct answer is: {answers}")
            total_questions += 1  # Increment the total questions for each question
        if total_questions > 0:
            print(f"\nYour total score: {count}/{total_questions}")
            print(f"Your score: {(count / total_questions) * 100:.2f}%")
        else:
            print("No Questions were found.")
        choice = input("Do you want to play again? (yes/no): ").strip().lower()
        if choice == 'yes':
            self.scrambled_words()
        else:
            print("Thank you for playing!")

# Example to run the game
B = BrainTeaser()
B.riddle()
B.scrambled_words()
```

```python
# Task 2
import random
class BRAINTEASER:
    def __init__(self):
        print("Welcome to the Game")
    def computer(self):
        count = 0
        total_questions = 0
        while True: 
            operator = random.choice(["+", "-", "*", "/"])
            a = random.randint(1, 20)
            b = random.randint(1, 20)
            if operator == "+":
                answer = a + b
                print(f"What is {a} + {b}?")
            elif operator == "-":
                answer = a - b
                print(f"What is {a} - {b}?")
            elif operator == "*":
                answer = a * b
                print(f"What is {a} * {b}?")
            elif operator == "/":
                if b != 0:
                    answer = a / b
                    print(f"What is {a} / {b}?")
                else:
                    print("Division by zero is not allowed.")
                    continue  
            user_answer = input("Enter your answer: ")
            try:
                if float(user_answer) == answer:
                    print("Correct!")
                    count += 1
                else:
                    print(f"Incorrect. The correct answer is: {answer}")
            except ValueError:
                print("Please enter a valid number.")
                continue
            total_questions += 1 
            choice = input("Do you want to continue playing? (Yes or No): ").strip().lower()
            if choice != 'yes':
                break  
        if total_questions > 0:
            print(f"\nYour total score: {count}/{total_questions}")
            print(f"Your score: {(count / total_questions) * 100:.2f}%")
            print("-----------------------------------------------")
        else:
            print("No questions were found.")
B = BRAINTEASER()
B.computer()
```

```python
# Task 3
class GeneralKnowledgeQuiz:
    def __init__(self):
        self.score = 0
        self.total_questions = 0
        self.quiz_data = {
            "Science": {
                "Which gas do plants absorb during photosynthesis?": "Carbon dioxide",
                "What is the full name of WWW?": "World Wide Web",
                "How many bones are there in the adult human body?": "206"
            },
            "Mathematics": {
                "What is the only even prime number?": "2",
                "What is the sum of the first 5 natural numbers?": "15",
                "What is the name of the mathematical symbol π?": "Pi"
            },
            "Geography": {
                "What is the tallest mountain on Earth?": "Mount Everest",
                "What is the capital of France?": "Paris",
                "What planet is known as the 'Red Planet'?": "Mars"
            },
            "Others": {
                "Which country won the most recent FIFA World Cup (2022)?": "Argentina",
                "Who invented the light bulb?": "Thomas Edison",
                "In what year did India declare itself a democratic republic state with the adoption of the constitution?": "1950"
            }
        }
    def ask_category(self):
        category=("Science","Mathematics","Geography","others")
        category = input("Enter the category: ").capitalize().strip()
        return category
    def ask_question(self, category):
        if category not in self.quiz_data:
            print("Invalid category! Please choose a valid category.")
            return False
        questions = self.quiz_data[category]
        for question, correct_answer in questions.items():
            print(f"\nQuestion: {question}")
            try:
                user_answer = input("Your answer: ").strip()
            except Exception as e:
                print(f"Error: {e}")
                continue 
            if self.check_answer(user_answer, correct_answer):
                print("Correct!")
                self.score += 1
            else:
                print(f"Incorrect! The correct answer is: {correct_answer}")
            self.total_questions += 1
        return True
    def check_answer(self, user_answer, correct_answer):
        return user_answer.lower() == correct_answer.lower()
    def display_score(self):
        if self.total_questions > 0:
            print(f"\nYour total score: {self.score}/{self.total_questions}")
            print(f"Your score: {(self.score / self.total_questions) * 100:.2f}%")
        else:
            print("No questions were answered.")
    def reset_score(self):
        self.score = 0
        self.total_questions = 0
    def play_game(self):
        while True:
            play_more = input("\nDo you want to start the quiz? (yes/no): ").strip().lower()
            if play_more == 'yes':
                self.reset_score()  
                category = self.ask_category()
                if self.ask_question(category):  
                    self.display_score()
            elif play_more == 'no':
                print("Thank you for playing! Goodbye!")
                break
            else:
                print("Invalid input. Please enter 'yes' or 'no'.")
if __name__ == "__main__":
    A=GeneralKnowledgeQuiz()
    A.play_game()
```
