```python
# Task 6
round2_qualifiers = pd.DataFrame({
    "Physics": Series_for_physics,
    "Maths": Series_for_maths
})[
    (Series_for_physics >= 50) & 
    (Series_for_maths >= 50) & 
    ((Series_for_physics + Series_for_maths) >= 120)
]
round2_qualifiers.reset_index(drop=True, inplace=True)
round2_qualifiers
```
